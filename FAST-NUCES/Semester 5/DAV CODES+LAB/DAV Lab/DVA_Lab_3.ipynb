```python
def detect_useless_col(df):
    null_percentage= (df.isna().sum() / len(df))*100
    useless = null_percentage[null_percentage > 60].index.tolist()
    return useless

useless_list = detect_useless_col(merge_df)
print(useless_list)
merge_df.drop(columns=useless_list)
```
