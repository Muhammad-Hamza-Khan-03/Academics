```python
import numpy as np

# Set the number of data points and the radius of the circles
n_samples = 100
radius1 = 1
radius2 = 2
radius3 = 3

# Generate data points for the first cluster
theta1 = np.linspace(0, np.pi/2, n_samples)
x1 = radius1 * np.cos(theta1) - radius3/2
y1 = radius1 * np.sin(theta1) - radius3/2

# Add some noise to the data points
x1 += np.random.normal(scale=0.1,size=n_samples)
y1 +=np.random.normal(scale=0.1,size=n_samples)

# Generate data points for the second cluster
theta2 = np.linspace(0,np.pi/2,n_samples)
x2=radius2*np.cos(theta2)-radius3/2
y2=radius2*np.sin(theta2)-radius3/2

# Add some noise to the data points
x2 +=np.random.normal(scale=0.01,size=n_samples)
y2 +=np.random.normal(scale=0.1,size=n_samples)

# Generate data points for the third cluster
theta3=np.linspace(0,np.pi/2,n_samples)
x3=radius3*np.cos(theta3)-radius3/2
y3=radius3*np.sin(theta3)-radius3/2

# Add some noise to the data points
x3+=np.random.normal(scale=0.01,size=n_samples)
y3+=np.random.normal(scale=0.1,size=n_samples)

X=np.concatenate([np.column_stack((x,y)) for x,y in zip([x1,x2,x3],[y1,y2,y3])])
Y=np.concatenate([[i]*n_samples for i in range(3)])

plt.scatter(X[:, 0], X[:, 1], c=Y,cmap='viridis')
plt.show()
```
